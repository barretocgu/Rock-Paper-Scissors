```python
from numpy.core.multiarray import result_type
import numpy as np
import random as rand

print("Hello! Welcome to a simulation of a rock, paper, scissors tournament (but with a twist). You will be asked if you want to name the players of your simulation. \n\
If you choose 'Y' you will be promted to enter the names of your players. Each round players will roll numbers between 1-10. If they happen to roll a '1' then the \n\
player will receive the advantage of an automatic win. If both players playing against each other receive the advantage then the round will play normally.\n")

# FUNCTIONS

# Creates an empty dictionary for all possible luck values 0-10 as keys and empty lists as values
def create_empty_dict10():
  empty_dict = dict()
  for value in range(0,11):
    empty_dict[value] = []
  return empty_dict

def assign_luck(num_players):
  luck_vals = []
  for luck_val in range(0,num_players):
      luck_vals.append(np.random.randint(11))
  return luck_vals


# Given a dictionary of luck values as keys  it will determine if any players got lucky and got a value of '1'
# Returns a list of lucky winners if there are any, else it returns an empty list
# Prints message
def check_winner(luck_dict):
  winner_list = []
  if not luck_dict[1] == []:
    print("\nCongratulations!: ")
    for winner in luck_dict[1]:
      winner_list.append(winner)
      print(winner)
    return winner_list
  else:
    "\nI'm sorry, there are no winners this time D: "
    return []

# Randomly determines the outcome of either Rock, Paper, or Scissors
# Returns outcome as string
def rock_paper_scissors():
  rand_val = np.random.randint(4)
  if rand_val == 0:
    return 'rock'
  elif rand_val == 1:
    return 'paper'
  else:
    return 'scissors'

# Given 2 players (p1 and p2) and two turn outcomes (t1 and t2) this function will determine a winner and return that winner
# If it is a draw it returns -1
def who_wins(p1, p2, t1, t2):
  if t1 == "rock":
    if t2 == "paper":
      return p2
    elif t2 == "scissors":
      return p1
  if t1 == "paper":
    if t2 == "scissors":
      return p2
    elif t2 == "rock":
      return p1
  else:
    if t2 == "rock":
      return p2
    elif t2 == "paper":
      return p1
  return -1

# This function simulates an instance of a game given the randomized game list and a list of luck winners
def play_game(name_list, luck_list):
  winner_list = []
  for player in range(0,len(name_list),2):
    p1 = name_list[player]
    p2 = name_list[player+1]
    if p1 in luck_list or p2 in luck_list:
      if p1 in luck_list and p2 in luck_list:
        print("Uh oh, both luck means no luck so let the games proceed!")
        winner_list.append(play_game([p1,p2],[])[0])
      elif p1 in luck_list:
        print(p1 + " has luck on their side and moves on!")
        print("Unfortunately that means that " + p2 + " is eliminated.\n")
        winner_list.append(p1)
      else:
        print(p2 + " has luck on their side and moves on!")
        print("Unfortunately that means that " + p1 + " is eliminated.\n")
        winner_list.append(p2)
    else:
      p1_move = rock_paper_scissors()
      p2_move = rock_paper_scissors()
      result = who_wins(p1,p2,p1_move,p2_move)
      print(p1 + ": "+ p1_move)
      print(p2 + ": "+ p2_move)
      if result != -1:
        print("Winner: " + result + "!!\n")
        winner_list.append(result)
      else:
        print("It's a draw so let's redo it!")
        winner_list.append(play_game([p1,p2],luck_list)[0])
  return winner_list


# Initiates by asking user for an input if they want to name their players
names = input("Would you like to name your players? (Y/N) ")

# Populates a list of names either by generic player names (i.e. 'Player 1') or user inputs
name_list = []
for name in range(1,9):
  if names == "N":
    name_list.append("Player " + str(name))
  else:
    name_list.append(input("Name of Player "+ str(name)+ ": "))

# Randomizes the pairings in the round and in the background makes a list of luck values to assign later
print("\nOk! We will randomize pairings. Here are the pairs for round 1:\n")
r1_names = name_list.copy()
rand.shuffle(r1_names)

r1_luck = assign_luck(8)

for r1 in range (0,8,2):
  print(str(r1_names[r1])+" vs "+str(r1_names[r1+1]))


# Waits for user input
input("\nPress 'enter' to proceed to luck draw: \n")

# Creates a dictionary called luck_draw and iterates through the players to print out their luck values and adds the player names
# as values to the keys which are the luck_numbers
luck_dict = create_empty_dict10()
for r1_luckdraw in range(0,8):
  print(str(name_list[r1_luckdraw]) + "'s Number: "+ str(r1_luck[r1_luckdraw]))
  luck_dict[r1_luck[r1_luckdraw]] += [name_list[r1_luckdraw]]

# Calls function to determine who has the lucky number
winners = check_winner(luck_dict)

# Waits for user input
input("\nPress 'enter' to proceed to round 1 results: \n")

# Will call the play_game function to simulate a round and will print out the winners at the end
win_list = play_game(r1_names, winners)
def printWinners(round_num, winner_list):
  print("The winners for round "+ str(round_num) + " are: ")
  for winner in winner_list:
    print(winner)
printWinners(1, win_list)

input("\nLet's move onto round 2! Press 'enter' to show the the pairings: \n")
r2_names = win_list.copy()
rand.shuffle(r2_names)

r2_luck = assign_luck(4)
for r2 in range (0,4,2):
  print(str(r2_names[r2])+" vs "+str(r2_names[r2+1]))

# Waits for user input
input("\nPress 'enter' to proceed to luck draw: \n")


luck_dict_r2 = create_empty_dict10()
for r2_luckdraw in range(0,4):
  print(str(win_list[r2_luckdraw]) + "'s Number: "+ str(r2_luck[r2_luckdraw]))
  luck_dict_r2[r2_luck[r2_luckdraw]] += [r2_names[r2_luckdraw]]

# Calls function to determine who has the lucky number
winners = check_winner(luck_dict_r2)

# Waits for user input
input("\nPress 'enter' to proceed to round 2 results: \n")

win_list = play_game(r2_names, winners)
printWinners(2, win_list)


input("\nFinal Round!! Press enter to show the final pairing:\n")

print(win_list[0] + " vs " + win_list[1])

flag = input("\nSince this is the final pairing we will make the luck draw optional. So you want to do a luck draw? (Y/N): ")

r3_names = win_list.copy()
if flag == "Y":
  r3_luck = assign_luck(2)
  luck_dict_r3 = create_empty_dict10()
  for r3_luckdraw in range(0,2):
    print(str(win_list[r3_luckdraw]) + "'s Number: "+ str(r3_luck[r3_luckdraw]))
    luck_dict_r3[r3_luck[r3_luckdraw]] += [r3_names[r3_luckdraw]]
  winners = check_winner(luck_dict_r3)
else:
  winners = []

input("\nPress 'enter' to see the results of the final round: \n")

win_list = play_game(r3_names, winners)

input("\nAnd the winner is..... (plz press 'enter'):\n")
print(win_list[0] + "!!!!!!!!\n")

rating = input("Yo Franklin rate this simulation on a score from 1-10.")

print("\nThank you for playing!")


```
